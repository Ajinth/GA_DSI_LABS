### 3. Set up the `Rank` variable as your target. How many classes are there?

```python
nhl['Rank'].value_counts()
```

```python
y = nhl['Rank']
```

We see that there are three categories of ranks. In other words there are three classes 

### 4. What is the baseline accuracy?

```python
number_1_rank = np.sum(nhl['Rank'] == 1)
number_2_rank = np.sum(nhl['Rank'] == 2)
number_3_rank = np.sum(nhl['Rank'] == 3)

baseline_accuracy_for_rank1 = number_1_rank/(len(nhl.Rank))
baseline_accuracy_for_rank2 = number_2_rank/(len(nhl.Rank))
baseline_accuracy_for_rank3 = number_3_rank/(len(nhl.Rank))

print ('Baseline Accuracy to Get Rank 1: ', baseline_accuracy_for_rank1)
print ('Baseline Accuracy to Get Rank 2: ', baseline_accuracy_for_rank2)
print ('Baseline Accuracy to Get Rank 3: ', baseline_accuracy_for_rank3)
```

### 5. Choose 4 features to be your predictor variables and set up your design matrix.

```python
# features_columns = ['GA', 'SA', 'FA', 'CA']
features_columns = ['CF%', 'GF', 'Sh%', 'PDO']

X = nhl[features_columns]
X.head()
```

```python
X.shape
```

### 6. Fit a `KNeighborsClassifier` with 1 neighbor using the target and predictors.

```python
from sklearn.neighbors import KNeighborsClassifier
knn_1 = KNeighborsClassifier(n_neighbors=1, weights = 'uniform')
knn_1.fit(X,y)
```

### 7. Evaluate the accuracy of your model.
- Is it better than baseline?
- Is it legitimate?

```python
'''Evaluating the Score of the Model'''

'''Calculating the Predicted Value'''
y_predict = knn_1.predict(X)


'''Validating the Score with the Confusion Matrix'''
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, y_predict)
cm
```

In other the matrix above tells us that our model predicts correctly 100% of the times. This is the case of overfitting  

```python
'''Validating the Score with the Model Score'''
from sklearn import metrics
print (metrics.accuracy_score(y_predict, y))
```

The score indicates that the model predicts the correcrt target variable 100% of the times.

### 8. Create a 50-50 train-test-split of your target and predictors. Refit the KNN and assess the accuracy.

We see that there is 50-50 split of the training and the test data

```python
'''Fitting with the Training Data'''
knn1_training_fit = knn_1.fit(X_train, y_train)
```

```python
'''Predicting Y Train with the Fitted Model'''
y_train_predicted = knn1_training_fit.predict(X_train)
```

```python
'''Score Evaluation with the Confusion Matrix'''
from sklearn.metrics import confusion_matrix
cm_train_score = confusion_matrix(y_train_predicted, y_train)
cm_train_score
```

### 11. Fit KNN across different values of K and plot the mean cross-validated accuracy with 5 folds.

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

fold_neighbors = int((X.shape[0] - X.shape[0]/5))

mean_cross_validated_score = [] 

for i in range(1, int(fold_neighbors)): 
    knn = KNeighborsClassifier(n_neighbors=i)
    score = np.mean(cross_val_score(knn, X, y, cv=5))
    mean_cross_validated_score.append(score)


fig, ax = plt.subplots(figsize = (10,10))
ax.plot(list(range(1, int(fold_neighbors))), mean_cross_validated_score,lw=3, color='g')    
```
