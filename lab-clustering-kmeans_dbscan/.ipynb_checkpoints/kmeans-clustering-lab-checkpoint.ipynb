### 3. Prepare the data for clustering

1. Remove the `species` column. We will see if the clusters from K-Means end up like the actual species.
2. Put the features on the same scale.

```python
# A:
from sklearn.preprocessing import StandardScaler

'''Using Standard Scalar to put the features on the same scale'''
ss = StandardScaler()
seeds_new = seeds.drop('species', axis=1)
seeds_new.head()

seeds_new_transform = ss.fit_transform(seeds_new)
seeds_new_transform[0:5]
```

### 4. Clustering with K-Means

- Cluster the data to our our target groups.
- We know that there are 3 actual classes. However, in an actual situation in which we used clustering we would have no idea. Lets initally try using the default K for `KMeans`(8).

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 8, n_jobs =-1)
km_fit = km.fit(seeds_new_transform)
km_fit.predict(seeds_new_transform)
```

### 5. Get the labels and centroids for out first clustering model.

```python
# A:
labels = km.labels_
centroids = km.cluster_centers_

print (labels)
print ('\n')
print (centroids)
```

### 6. Compute the silouette score and visually examine the results of the 8 clusters. 

_(pairplot with hue)_

```python
from sklearn.metrics import silhouette_score


print ('S Score with the Transformed Data', silhouette_score(seeds_new_transform, km_fit.labels_))
print ('S Score with the  Non Transformed Data', silhouette_score(seeds_new, km_fit.labels_))
```

We see that the model performs marginally better on the Standardized data as compared to the Non Standardized Data
