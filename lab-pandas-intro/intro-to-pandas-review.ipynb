#### 4. [Rename any columns](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.rename.html) with spaces or special characters to not contain any.

```python
df.rename(columns = {'Show Start': 'Show_Start'}, inplace=True)
df.rename(columns = {'Show End': 'Show_End'}, inplace=True)
df.rename(columns = {'Status?': 'Status'}, inplace=True)
df.rename(columns = {'Years Since': 'Years_Since'}, inplace=True)
df.rename(columns = {'#LEAD': 'Lead'}, inplace=True)
df.rename(columns = {'#SUPPORT': 'Support'}, inplace=True)
df.rename(columns = {'#Shows': 'Shows'}, inplace=True)
df.rename(columns = {'Score/Y': 'Score_Y'}, inplace=True)
```

```python
# df_no_end = ~df['Status'].str.contains('End')
# df.loc[df_no_end, :]

df.loc[(df['Status']!='END')& (df['Status']!='End'), :]
```

**13. Which day of the week has the most crime?**

```python
sf.head()
# sf.groupby('DayOfWeek')
```

```python
day_of_week = sf.groupby('DayOfWeek').count()
sorted_day_of_week = day_of_week['Category'].sort_values(ascending=False)
sorted_day_of_week[:1]
```

#### 14. Make a new DataFrame featuring the crime categories and the number of crimes per category.

```python
crime_cat = sf.groupby('Category').count()

crime_category_count = crime_cat['Descript'].to_frame(name=None)
crime_category_count.rename(columns = {'Descript': 'Count of Crimes'}, inplace=True)
crime_category_count
```

```python
sf.head()
```

**15. Make a DataFrame that includes the districts and crime counts per district. Which district has the most crime?**

*Hint: You can use the `.sort_values()` function to sort your DataFrame by column.*

```python
district_df = sf.groupby('PdDistrict').count()
district_crime_count = district_df['Dates'].sort_values(ascending=False)
district_crime_count_df = district_crime_count.to_frame()
district_crime_count_df.rename(columns = {'Dates': 'Count' }, inplace=True)
```
