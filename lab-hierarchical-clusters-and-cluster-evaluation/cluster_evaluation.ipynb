```python
import pylab 

'''Defining a basic function to plot the Scatter plots based on color coded labels'''

def basic_plotter(df, title):
    fig, ax  = plt.subplots(figsize=(10,10))
    plt.scatter(df['x'], df['y'], c=df.label, cmap=pylab.cm.jet)
    plt.title(title,fontsize=25)
    plt.show()

'''Calling the plotter function to plot the graphs for the different datasets'''

datasets = [flame , agg, comp, jain , path , r15, spiral]     
datasets_title = ['flame' , 'agg', 'comp', 'jain' , 'path' , 'r15', 'spiral']
for df, title in zip(datasets, datasets_title): 
    basic_plotter(df,title)
```

```python
km = KMeans(n_clusters = 2)
km.fit(flame.loc[:, ['x','y']], flame.label)
predicted_km_labels = km.labels_
```

```python
ag = AgglomerativeClustering(n_clusters = 2)
ag.fit(flame.loc[:, ['x','y']], flame.label)
predicted_ag_labels = ag.labels_
```

```python
db = DBSCAN(eps=0.5, min_samples=5)
db.fit(flame.loc[:, ['x','y']], flame.label)
predicted_db_labels = db.labels_
```

```python
datasets = [flame , agg, comp, jain , path , r15, spiral]     
datasets_title = ['flame' , 'agg', 'comp', 'jain' , 'path' , 'r15', 'spiral']

'''Function to define the plotter '''
def clustering_plotter(df, title, km_nclusters, ag_nclusters, db_eps,  db_min_samples): 
    
    '''Handling the KMeans Fit and Plot'''
    km = KMeans(km_nclusters)
    km.fit(df.loc[:, ['x','y']], df.label)
    predicted_km_labels = km.labels_
    
    fig, ax  = plt.subplots(figsize=(6,6))
    plt.scatter(df['x'], df['y'], c=predicted_km_labels, cmap=pylab.cm.jet)
    plt.xlabel(df.columns[0],fontsize=25)
    plt.ylabel(df.columns[1],fontsize=25)
    plt.title('K-Means'+ '-' + title,fontsize=25)
    plt.show()
    print ('---------------------------------------------------------------------------------------------------')
    
    
    '''Handling the Agglomerative Fit and Plot'''
    am = AgglomerativeClustering(km_nclusters)
    am.fit(df.loc[:, ['x','y']], df.label)
    predicted_am_labels = am.labels_
    
    fig, ax  = plt.subplots(figsize=(6,6))
    plt.scatter(df['x'], df['y'], c=predicted_am_labels, cmap=pylab.cm.jet)
    plt.xlabel(df.columns[0],fontsize=25)
    plt.ylabel(df.columns[1],fontsize=25)
    plt.title('Agglomerative'+ '-' + title,fontsize=25)
    plt.show()
    print ('----------------------------------------------------------------------------------------------------')
    
    
    '''Handling the DBScan Fit and Plot'''
    db = DBSCAN(db_eps,db_min_samples)
    db.fit(df.loc[:, ['x','y']], df.label)
    predicted_db_labels = db.labels_
    
    fig, ax  = plt.subplots(figsize=(6,6))
    plt.scatter(df['x'], df['y'], c=predicted_db_labels, cmap=pylab.cm.jet)
    plt.xlabel(df.columns[0],fontsize=25)
    plt.ylabel(df.columns[1],fontsize=25)
    plt.title('DBSCAN'+ '-' + title,fontsize=25)
    plt.show()
    print ('====================================================================================================')
    

    
'''Wrapper function to call the plotter'''

def plot_initiator(): 
    for df, title in zip(datasets, datasets_title): 
        clustering_plotter(df,title,2,2,0.5,5)

'''Call the plot initiator'''

plot_initiator()  
        

        
# clustering_plotter(agg,2,2,0.5,5)
```

## 5. [Bonus] Explore some other clustering algorithms

sklearn comes with a variety of unsupervised clustering algorithms, some of which we have not covered in class. Two algorithms in particular may be of interest to you:

1. [Affinity Propagation](http://scikit-learn.org/dev/modules/clustering.html#affinity-propagation) finds clusters by "sending messages" from points to other points. Points group into clusters based on a "damping factor". The main appeal of affinity propagation is that the number of clusters do not need to be specified by the user (like DBSCAN).
- [Birch](http://scikit-learn.org/dev/modules/clustering.html#birch) finds clusters with a tree-based algorithm (somewhat) reminiscent of decision trees. It finds clusters by evaluating branches/nodes on a tree that best describe the features of the data.
